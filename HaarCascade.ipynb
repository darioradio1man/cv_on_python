```python
import pandas
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
import random 
import time
%matplotlib inline

df = pandas.read_csv('./good.csv')
df.path
img = cv2.imread(df.path[0], 1)
plt.imshow(img,cmap='brg')
plt.show()

x, y, w, h = df.xMin[0], df.yMin[0], (df.xMax[0] - df.xMin[0]), (df.yMax[0] - df.yMin[0])
print(x, y, w, h)
print(df.xMin[0], df.yMin[0], df.xMax[0], df.yMax[0])
print((df.xMax[0] - df.xMin[0])* (df.yMax[0] - df.yMin[0]))

roi = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),-1)
plt.imshow(roi,cmap='gray')
plt.show()
random.seed(time.clock())
flags = 0
for i in range(1, w):
    for j in range(1,h):
        if ((i * j) < int(0.5 * w * h)):
            flags+=1
            roi = cv2.rectangle(img, (random.randint(0, np.size(img, 1)-1) , random.randint(0, np.size(img, 0)-1)), (random.randint(0,i), random.randint(0, j)), (255,0,0),-1)
            plt.imshow(roi,cmap='gray')
            plt.show()
        else:
            break
print(flags)
```

```python
import pandas
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
import random 
import time
%matplotlib inline

df = pandas.read_csv('./good.csv')
df.path

x, y, w, h = df.xMin[0], df.yMin[0], (df.xMax[0] - df.xMin[0]), (df.yMax[0] - df.yMin[0])
print(" : " + str(x) + ", " + str(y))
print(" : " + str(x+w) + ", " + str(y))
print(" : " + str(x) + ", " + str(y+h))
print(" : " + str(x+w) + ", " + str(y+h))
pos = cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),-1)
plt.imshow(pos,cmap='gray')
plt.show()

x1 = random.randint(0, img.shape[1])
y1 = random.randint(0, img.shape[0])
w1 = random.randint(0, img.shape[1])
h1 = random.randint(0, img.shape[0])
if((x1 + w1 < x) & (y1 + h1 < y)):
    neg = cv2.rectangle(img, (x1 , y1), (x1+w1, y1+h1), (255,0,0),-1)
    plt.imshow(neg,cmap='gray')
    plt.show()
```
